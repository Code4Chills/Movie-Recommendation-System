```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data=pd.read_csv('C:/Users/Pearry/Downloads/datasets/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

feature selection


```python
selected_features=['genres','keywords','tagline','cast','director']
```

```python
for feature in selected_features:
    movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combining all 5 selected features

combined_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

**Cosine Similarity**

```python
#getting similarity scores using cosine similarity
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
#getting movie name from users
movie_name=input('Enter your favorite movie name:')
```

```python
 #creating list with all movie names given in dataframe
    
movies_list=movies_data['title'].tolist()
print(movies_list)
```

```python
#finding close match to movie name input by user
find_close_match=difflib.get_close_matches(movie_name,movies_list)
print(find_close_match)
```

```python
closest_match=find_close_match[0]
print(closest_match)
```

```python
#finding the index of the movie
movie_index=movies_data[movies_data.title==closest_match]['index'].values[0]
```

```python
print(movie_index)
```

```python
#getting a list of similar movies
similarity_score=list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on index
print('Movies Suggested for you:')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if i<20:
        print(i,'.',title_from_index)
        i+=1
```
